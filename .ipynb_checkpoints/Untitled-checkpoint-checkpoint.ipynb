```python
# Drop Serial No. and clean column names
df = df.drop('University Rating', axis=1)
df = df.rename(columns={
    'LOR ': 'LOR',
    'Chance of Admit ': 'Admission_Chance'
})
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Outlier Detection")
plt.show()
```

```python
from scipy.stats.mstats import winsorize

# Updated winsorization configuration with all features
winsorize_config = {
    'LOR': (0.05, 0.05),          # Letters of Recommendation (more variable)
    'SOP': (0.05, 0.05),          # Statement of Purpose
    'CGPA': (0.01, 0.01),         # Cumulative GPA (more critical)
    # 'University Rating': (0.03, 0), # Only cap upper outliers
    'Admission_Chance': (0.01, 0.01),
    'Research': None              # Binary feature - no winsorization
}

# Create list of columns from config for plotting
winsorize_cols = list(winsorize_config.keys())

# Preserve original data before modification
df_original = df.copy()

# Apply winsorization efficiently
for col, limits in winsorize_config.items():
    df[col] = winsorize(df[col], limits=limits)
```

```python
# Improved visualization with all features
plt.figure(figsize=(18, 8))

# Before treatment
plt.subplot(1, 2, 1)
sns.boxplot(data=df_original[winsorize_cols], palette="viridis")
plt.title("Pre-Treatment Distributions", fontsize=14)
plt.xticks(rotation=30)

# After treatment 
plt.subplot(1, 2, 2)
sns.boxplot(data=df[winsorize_cols], palette="plasma")
plt.title("Post-Winsorization Distributions", fontsize=14)
plt.xticks(rotation=30)

plt.suptitle("Outlier Treatment Comparison", y=1.05, fontsize=16)
plt.tight_layout()
plt.show()
```

Outlier Detection and Treatment
Boxplots revealed a few mild outliers in features like LOR, CGPA, and Chance of Admit. Rather than removing them, which can cause information loss, we applied Winsorization at the 1% level to cap the extreme values. This method maintains dataset size while controlling for skewed influence.

## 4. Feature Selection 

```python
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression

# Prepare data
X = df.drop(['Admission_Chance', 'Total_Score'], axis=1)
y = df['Admission_Chance']
```

```python
# 1. Correlation Analysis 
print("Feature Correlation with Target:")
corr_scores = X.corrwith(y).abs().sort_values(ascending=False)
print(corr_scores.to_string(float_format="%.3f"))
```

```python
# 2. Statistical Significance (F-Test)
print("\nF-Regression Scores:")
selector = SelectKBest(f_regression, k='all').fit(X, y)
f_scores = pd.DataFrame({
    'Feature': X.columns,
    'F-Score': selector.scores_,
    'p-value': selector.pvalues_
}).sort_values('F-Score', ascending=False)
print(f_scores.to_string(float_format="%.3f"))
```

```python
# 3. Mutual Information (Non-linear Relationships)
print("\nMutual Information Scores:")
mi_scores = pd.Series(
    mutual_info_regression(X, y),
    index=X.columns
).sort_values(ascending=False)
print(mi_scores.to_string(float_format="%.3f"))
```

```python
# 4. Final Feature Selection (Consensus Approach)
feature_rankings = pd.DataFrame({
    'Correlation': corr_scores,
    'F-Score': f_scores.set_index('Feature')['F-Score'],
    'MI': mi_scores
})
```

```python
# Normalize scores (0-1) for comparison
feature_rankings = feature_rankings.apply(lambda x: x/x.max(), axis=0)

# Calculate composite score (weighted average)
feature_rankings['Composite_Score'] = (
    0.4 * feature_rankings['Correlation'] + 
    0.4 * feature_rankings['F-Score'] + 
    0.2 * feature_rankings['MI']
)

print("\n Final Feature Rankings:")
print(feature_rankings.sort_values('Composite_Score', ascending=False).to_string(float_format="%.3f"))
```

```python
# Select top features (adjust threshold as needed)
selected_features = feature_rankings[feature_rankings['Composite_Score'] > 0.6].index.tolist()
print("\n Selected Features:", selected_features)
```

## 5. Model Development

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
# Prepare data
X = df.drop(['Admission_Chance', 'Total_Score'], axis=1)
y = df['Admission_Chance']
```

```python
# Split Data into Train/Test Sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,  # 20% for testing
    random_state=42  # For reproducibility
)
```

```python
# Scale the Features (Important for some models)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform train data
X_test_scaled = scaler.transform(X_test) 
```

```python
# Define evaluation function
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # Cross-validation
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='r2')
    
    print(f"\n{model_name} Performance:")
    print(f"R Score: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"Cross-validated R: {np.mean(cv_scores):.4f} ({np.std(cv_scores):.4f})")
    
    return {
        'model': model_name,
        'r2': r2,
        'mae': mae,
        'rmse': rmse,
        'cv_r2': np.mean(cv_scores)
    }
```

```python
svr = SVR()
rf = RandomForestRegressor(random_state=42)
gbdt = GradientBoostingRegressor(random_state=42)
knn = KNeighborsRegressor()
rr = Ridge(alpha=1.0)
lf = LinearRegression()
```

```python
models = {
    'Linear Regression': lf,
    'Ridge Regression': rr,
    'Random Forest': rf,
    'Gradient Boosting': gbdt,
    'SVR': svr,
    'KNN': knn,
}    
```

```python
# Evaluate all models
results = []
for name, model in models.items():
    results.append(evaluate_model(model, X_train_scaled, X_test_scaled, y_train, y_test, name))
```

## 6. Hyperparameter Tuning

```python
optimized_param_grid = {
    'n_estimators': [150, 200, 250],  # More focused around best expected value
    'max_depth': [15, 20, 25],        # Removed None to prevent overfitting
    'min_samples_split': [2, 3, 5],   # More granular low values
    'min_samples_leaf': [1, 2],       # Removed 4 (often too restrictive)
    'max_features': ['sqrt', 'log2'],  # Removed None
    'bootstrap': [True, False]        # Added important new parameter
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Initialize with warm_start=True for faster iterations
rf = RandomForestRegressor(random_state=42, warm_start=True)

# Improved GridSearch configuration
optimized_search = GridSearchCV(
    estimator=rf,
    param_grid=optimized_param_grid,
    cv=5,   
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    return_train_score=True  # Helps diagnose overfitting
)
```

```python
# Fit the model
optimized_search.fit(X_train_scaled, y_train)

# Best model analysis
best_rf = optimized_search.best_estimator_
print("Best Parameters:", optimized_search.best_params_)
print("Best CV Score:", optimized_search.best_score_)
```

```python
# Original model
original_rf = RandomForestRegressor(random_state=42)
original_rf.fit(X_train_scaled, y_train)
original_score = original_rf.score(X_test_scaled, y_test)

# Tuned model
tuned_score = best_rf.score(X_test_scaled, y_test)

print(f"Original R: {original_score:.4f}")
print(f"Tuned R: {tuned_score:.4f}")
print(f"Improvement: {(tuned_score - original_score):.4f}")
```

```python
# Feature importance
importances = best_rf.feature_importances_
features = X_train.columns

# Create DataFrame
fi_df = pd.DataFrame({'Feature': features, 'Importance': importances})
fi_df = fi_df.sort_values('Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=fi_df, palette='viridis')
plt.title('Optimized Random Forest Feature Importance')
plt.show()
```

```python
from sklearn.model_selection import learning_curve

# Create the learning curve data
train_sizes, train_scores, test_scores = learning_curve(
    estimator=best_rf,  # Your optimized Random Forest model
    X=X_train_scaled,
    y=y_train,
    cv=5,  # Number of cross-validation folds
    scoring='r2',  # Evaluation metric
    n_jobs=-1,  # Use all processors
    train_sizes=np.linspace(0.1, 1.0, 10)  # 10 different training set sizes
)

# Now you can plot the learning curve
plt.figure(figsize=(10,6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(test_scores, axis=1), 's--', label='Validation')
plt.fill_between(train_sizes, 
                 np.mean(train_scores, axis=1) - np.std(train_scores, axis=1),
                 np.mean(train_scores, axis=1) + np.std(train_scores, axis=1),
                 alpha=0.1)
plt.title('Learning Curve')
plt.xlabel('Training Examples')
plt.ylabel('R Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import validation_curve

param_range = [5, 10, 15, 20, 25, 30, None]

train_scores, val_scores = validation_curve(
    RandomForestRegressor(random_state=42),
    X_train_scaled, y_train,
    param_name="max_depth",
    param_range=param_range,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

train_scores_mean = np.mean(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)

plt.plot(param_range, train_scores_mean, label="Training score", marker='o')
plt.plot(param_range, val_scores_mean, label="Validation score", marker='o')
plt.xlabel("max_depth")
plt.ylabel("R Score")
plt.title("Validation Curve for Random Forest - max_depth")
plt.legend()
# plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
param_range = [10, 50, 100, 150, 200]  
param_name = "n_estimators"

train_scores, val_scores = validation_curve(
    RandomForestRegressor(random_state=42),
    X_train_scaled, y_train,
    param_name=param_name,
    param_range=param_range,
    cv=5,
    scoring="r2",
    n_jobs=-1
)

# Mean scores
train_mean = np.mean(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)


# Plot
plt.plot(param_range, train_mean, label='Train', marker='o')
plt.plot(param_range, val_mean, label='Validation', marker='o')
plt.xlabel(param_name)
plt.ylabel("R Score")
plt.title(f"Validation Curve for Random Forest - {param_name}")
plt.legend()
plt.tight_layout()
plt.show()
```
