## University Admission Prediction
Featuring various Machine Learning models to predict Master's university admission acceptance probability.

In this project, I have used various Machine Learning models to predict the chance of admission into a university based on the student's academic and personal profile.

### Project Objective
This project aims to build a system that can predict the probability of admission for students applying to graduate programs. By analyzing key application features, the system helps students understand their admission chances and universities identify strong candidates.

##  Machine Learning Models Used

I trained and evaluated several Machine Learning models to compare their efficiency and prediction accuracy:

1. **Support Vector Classifier (SVC) Model**  
   A classification model to predict whether a student is likely to get admitted or not.

2. **Random Forest Classifier Model**  
   A powerful ensemble model that builds multiple decision trees and aggregates their results.

3. **Logistic Regression Model**  
   A simple but effective model for binary classification problems like admission (Yes/No).

4. **K-Nearest Neighbors (KNN) Model**  
   A model that predicts outcomes based on the closest examples in the dataset.

##  Data Source

The dataset used in this project is partially based on the admissions dataset available on [Kaggle](https://www.kaggle.com/mohansacharya/graduate-admissions).  
To enrich the dataset and improve model performance, I created approximately 50% of the data entries myself by generating additional records based on realistic values.



##  Dataset Features

The dataset contains several parameters that are considered important during applications for graduate programs:

- **GRE Score** (out of 340)  
  Graduate Record Exam score.

- **TOEFL Score** (out of 120)  
  Test of English as a Foreign Language score.

- **University Rating** (out of 5)  
  Reputation/ranking of the university applied to.

- **Statement of Purpose (SOP)** strength (out of 5)  
  Quality of the SOP submitted.

- **Letter of Recommendation (LOR)** strength (out of 5)  
  Strength of recommendation letters submitted.

- **CGPA** (out of 10)  
  Undergraduate GPA.

- **Research Experience** (either 0 or 1)  
  Whether the student has prior research experience.

- **Chance of Admit** (ranging from 0 to 1)  
  Target variable â€” represents the likelihood of admission.

## 4. Feature Selection 
- Conducted relevance analysis through correlation metrics and feature importance scores.
- Removed redundant or irrelevant features to simplify the model.

```python
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression

# Prepare data
X = df.drop(['Admission_Chance', 'Total_Score'], axis=1)
y = df['Admission_Chance']
```

```python
X
```

```python
# 1. Correlation Analysis 
print("Feature Correlation with Target:")
corr_scores = X.corrwith(y).abs().sort_values(ascending=False)
print(corr_scores.to_string(float_format="%.3f"))
```

```python
# 2. Statistical Significance (F-Test)
print("\nF-Regression Scores:")
selector = SelectKBest(f_regression, k='all').fit(X, y)
f_scores = pd.DataFrame({
    'Feature': X.columns,
    'F-Score': selector.scores_,
    'p-value': selector.pvalues_
}).sort_values('F-Score', ascending=False)
print(f_scores.to_string(float_format="%.3f"))
```

```python
# 3. Mutual Information (Non-linear Relationships)
print("\nMutual Information Scores:")
mi_scores = pd.Series(
    mutual_info_regression(X, y),
    index=X.columns
).sort_values(ascending=False)
print(mi_scores.to_string(float_format="%.3f"))
```

```python
# 4. Final Feature Selection (Consensus Approach)
feature_rankings = pd.DataFrame({
    'Correlation': corr_scores,
    'F-Score': f_scores.set_index('Feature')['F-Score'],
    'MI': mi_scores
})
```

```python
# Normalize scores (0-1) for comparison
feature_rankings = feature_rankings.apply(lambda x: x/x.max(), axis=0)

# Calculate composite score (weighted average)
feature_rankings['Composite_Score'] = (
    0.4 * feature_rankings['Correlation'] + 
    0.4 * feature_rankings['F-Score'] + 
    0.2 * feature_rankings['MI']
)

print("\n Final Feature Rankings:")
print(feature_rankings.sort_values('Composite_Score', ascending=False).to_string(float_format="%.3f"))
```

```python
# Select top features (adjust threshold as needed)
selected_features = feature_rankings[feature_rankings['Composite_Score'] > 0.6].index.tolist()
print("\n Selected Features:", selected_features)
```

## 5. Model Development

In this section, we developed and evaluated multiple models for both Regression and Classification tasks using the Admission Prediction dataset.

Our goal was to:

Predict Chance of Admit using Regression models.

Classify candidates into Admit or Not Admit categories using Classification models.

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
# Prepare data
features = ['GRE_Score', 'TOEFL_Score', 'SOP', 'LOR', 'CGPA'] 
X = df[features] 
y = df['Admission_Chance']
```

```python
# Split Data into Train/Test Sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,  # 20% for testing
    random_state=42  # For reproducibility
)
```

```python
# Scale the Features (Important for some models)
scaler = StandardScaler()
```

```python

X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform train data
X_test_scaled = scaler.transform(X_test)
X_scaled = scaler.transform(X)
```

```python
scaler
```

##  Regression Models
We evaluated the following Regression models:

Linear Regression (LR)

Support Vector Regression (SVR)

Random Forest Regression (RF)

Gradient Boosting Regression (GR)

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor


# Initialize models
models = {
    "LR": LinearRegression(),
    "SVR": SVR(),
    "RF": RandomForestRegressor(random_state=42),
    "GR": GradientBoostingRegressor(random_state=42)
}

results = []
score_data = []

for model_name, model in models.items():
    # Train model
    model.fit(X_train_scaled, y_train)
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    
    # Calculate metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2')
    
    # Store results for table
    results.append({
        "Model": model_name,
        "RÂ² Score": r2,
        "MAE": mae,
        "RMSE": rmse,
        "CV RÂ² (Mean Â± Std)": f"{np.mean(cv_scores):.4f} Â± {np.std(cv_scores):.4f}"
    })
    
    # Store for visualization
    # score_data.append({
    #     "Models": model_name,
    #     "SCORE": r2
    # })
    
    # Print detailed performance
    print(f"\n{model_name} Performance:")
    print(f"RÂ² Score: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"Cross-validated RÂ²: {np.mean(cv_scores):.4f} (Â±{np.std(cv_scores):.4f})")

# Create dataframes
final_results = pd.DataFrame(results)
final_data = pd.DataFrame(score_data)
```

```python
# Display the metrics table
print("\nDetailed Model Performance:")
final_results
```

### Observations:
Random Forest Regression achieved the highest RÂ² score among all models, demonstrating strong predictive power with low MAE and RMSE.

Support Vector Regression performed comparably to Linear Regression with slightly better cross-validation RÂ² score.

LinearRegression also delivered competitive performance.

GradientBoostingRegressor performed well but was slightly behind other models.

## Classification Models
For the Classification task, we evaluated the following models:

Logistic Regression

Support Vector Classifier (SVC)

K-Nearest Neighbors (KNN)

Random Forest Classifier

Gradient Boosting Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                            f1_score, roc_auc_score, confusion_matrix, 
                            classification_report)
from sklearn.model_selection import cross_val_score
```

```python
# Convert to binary classification (1 if Chance of Admit > 0.8, else 0)
y_train = np.array([1 if value > 0.8 else 0 for value in y_train])
y_test = np.array([1 if value > 0.8 else 0 for value in y_test])

# Initialize models
models = {
    "LR": LogisticRegression(max_iter=1000),
    "SVC": SVC(probability=True),
    "K-Neighbors": KNeighborsClassifier(),
    "RF": RandomForestClassifier(random_state=42),
    "GB": GradientBoostingClassifier(random_state=42)
}

results = []
acc_data = []

for model_name, model in models.items():
    # Train model
    model.fit(X_train_scaled, y_train)
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:, 1] if hasattr(model, "predict_proba") else [0]*len(y_test)
    
    # Calculate metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba) if hasattr(model, "predict_proba") else np.nan
    
    # Cross-validation
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    
    # Store results
    results.append({
        "Model": model_name,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1 Score": f1,
        "ROC AUC": roc_auc,
        "CV Accuracy (Mean Â± Std)": f"{np.mean(cv_scores):.4f} Â± {np.std(cv_scores):.4f}"
    })
    
    acc_data.append({
        "Models": model_name,
        "ACC": acc
    })

# Create dataframes
final_results = pd.DataFrame(results)
final_data = pd.DataFrame(acc_data)

# Display classification reports
print("\nDetailed Classification Reports:")
for model_name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))
```

```python
## Accuracy Plot

plt.figure(figsize=(6, 6))
sns.set(style="whitegrid", palette="Set2") 

ax1 = sns.barplot(data=final_data, x="Models", y="ACC") 
ax1.set_title('Model Accuracy Comparison', fontsize=14)  
ax1.set_xlabel('')  
ax1.set_ylabel('Accuracy', fontsize=12)  
ax1.set_ylim(0.8, 1.0)

for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.3f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                fontsize=11, color='black', 
                xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
#  Metrics Comparison 
plt.figure(figsize=(8, 6))
metrics_df = final_results.melt(id_vars=['Model'], 
                              value_vars=['Precision', 'Recall', 'F1 Score'],
                              var_name='Metric', 
                              value_name='Score')

ax2 = sns.barplot(data=metrics_df, x='Model', y='Score', hue='Metric')
ax2.set_title('Precision, Recall and F1 Scores', fontsize=14)
ax2.set_xlabel('')
ax2.set_ylabel('Score', fontsize=12)
ax2.set_ylim(0.8, 1.0)
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
## SVC
param_grid_svc = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf']
}

svc_model = SVC(probability=True, random_state=42)

# 3. Fit the model on the training data
svc_model.fit(X_train_scaled, y_train)

# # 4. Predict on test set
# y_pred = svc_model.predict(X_test_scaled)

grid_svc = GridSearchCV(estimator=svc_model,
                        param_grid=param_grid_svc,
                        cv=5,
                        scoring='accuracy',
                        n_jobs=-1,
                        verbose=2)

grid_svc.fit(X_train_scaled, y_train)

print("SVC Best Parameters:", grid_svc.best_params_)
print("SVC Best CV Accuracy:", grid_svc.best_score_)

# Evaluate on test set
best_svc = grid_svc.best_estimator_
y_pred_svc = best_svc.predict(X_test_scaled)
y_proba_svc = best_svc.predict_proba(X_test_scaled)[:, 1]

print("\nSVC Final Performance on Test Set:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svc):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_svc):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_svc):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_svc):.4f}")
print(f"ROC AUC: {roc_auc_score(y_test, best_svc.predict_proba(X_test_scaled)[:,1]):.4f}")
```

```python
## Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

## Initialize Mode
rf_model = RandomForestClassifier(random_state=42)

## Grid Search
grid_rf = GridSearchCV(estimator=rf_model,
                       param_grid=param_grid_rf,
                       cv=5,
                       scoring='accuracy',
                       n_jobs=-1,
                       verbose=2)

## Fit the model on training dat
grid_rf.fit(X_train_scaled, y_train)

## Best Parameters
print("\nRF Best Parameters:", grid_rf.best_params_)
print("RF Best CV Accuracy:", grid_rf.best_score_)

## Final Evaluation on Test Set
best_rf = grid_rf.best_estimator_

# Predict
y_pred_rf = best_rf.predict(X_test_scaled)
y_proba_rf = best_rf.predict_proba(X_test_scaled)[:, 1]

print("\nRF Final Performance on Test Set:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_rf):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_rf):.4f}")
print(f"ROC AUC: {roc_auc_score(y_test, y_proba_rf):.4f}")
```
